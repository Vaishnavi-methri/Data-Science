```python
df = pd.read_csv("/content/student_sample_Cxsxsl6los (1).csv")
df
```

```python
# dat cleaning
# Fill the missingplace
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age,inplace=True)
df
```

```python
#Drop Duplicate
df.drop_duplicates(inplace=True)
df
```

```python
# Convert Age to integer
df['Age'] = df['Age'].astype(int)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#bar Chart for Grade Distribution
grade_counts = df['Grade'].value_counts()
plt.figure(figsize=(8,6))
sns.barplot(x=grade_counts.index,y=grade_counts.values,palette='viridis')
```

```python
# pie Chart For Gender Distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
```

```python
# Histogram for Age
plt.figure(figsize=(8,6))
sns.histplot(df['Age'],bins=10,kde=True,color='skyblue')
plt.hist(df['Age'],bins=10,color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
#for now lets simulate a score column
df['Score'] = np.random.randint(60,100,size=len(df))
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age',y='Score',hue='Gender',data=df,palette='coolwarm')
```
